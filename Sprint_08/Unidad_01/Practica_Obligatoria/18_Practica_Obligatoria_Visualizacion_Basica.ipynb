### #1.1

Carga el dataset de casas de California que está en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripción. Deshazte de las filas con nulos.

```python
df_california = pd.read_csv("./data/california_cities.csv")
df = df_california.copy()
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
ft_to_m = 0.3048
df['elevation_m'].fillna(df['elevation_ft'] * ft_to_m, inplace=True)
df.info()
```

```python
m_to_ft = 3.28084
df['elevation_ft'].fillna(df['elevation_m'] * m_to_ft, inplace=True)
df.info()
```

```python
mi_to_km2 = 2.58999
df['area_total_km2'].fillna(df['area_total_sq_mi'] * mi_to_km2, inplace=True)
df['area_land_km2'].fillna(df['area_land_sq_mi'] * mi_to_km2, inplace=True)
df['area_water_km2'].fillna(df['area_water_sq_mi'] * mi_to_km2, inplace=True)
df.info()
```

```python

km2_to_mi = 0.386102
df['area_total_sq_mi'].fillna(df['area_total_km2'] * km2_to_mi, inplace=True)
df['area_land_sq_mi'].fillna(df['area_land_km2'] * km2_to_mi, inplace=True)
df['area_water_sq_mi'].fillna(df['area_water_km2'] * km2_to_mi, inplace=True)
df.info()
```

```python
df["area_water_percent"].fillna(df['area_water_km2'] / df['area_total_km2'] * 100, inplace=True)
df.info()
```

```python
df["area_total_km2"].fillna(df["area_land_km2"]+df["area_water_km2"],inplace = True)
df["area_total_sq_mi"].fillna(df["area_land_sq_mi"]+df["area_water_sq_mi"],inplace = True)
df.info()
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center").
2. Histograma del área total ocupada por las ciudades.
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python


# Obtener las ciudades más pobladas de California
ciudades_mas_pobladas = df.sort_values(by='population_total', ascending=False).head(5)
colores_pastel = ['#4b54b3', '#999996', '#731e2b', '#4bb34b', '#4f3827']
# Crear el diagrama de barras
fig, ax = plt.subplots()
barras = ax.bar(ciudades_mas_pobladas['city'], ciudades_mas_pobladas['population_total'], color = colores_pastel)

for barra in barras:
    altura = barra.get_height()
    ax.annotate('{}'.format(altura),
                xy=(barra.get_x() + barra.get_width() / 2, altura),
                xytext=(0, 3),  # 3 puntos de desplazamiento vertical
                textcoords="offset points",
                ha='center', va='bottom')

# Configurar etiquetas y título
ax.set_title("Ciudades más pobladas de California")
ax.set_ylabel("Habitantes (miles)")
ax.set_ylim(0, 5000000)
ax.set_yticks([1000000, 2000000, 3000000, 4000000, 5000000]);




```

```python
font = {'family': 'Comic Sans MS', 'weight': 'normal', 'size': 12}

# Crear el histograma
plt.hist(df['area_total_km2'], bins=40, alpha=0.7, color="steelblue")

# Configurar etiquetas y título con el estilo de la fuente
plt.xlabel("Superficie (km2)", fontdict=font)
plt.ylabel("Número de ciudades", fontdict=font)
plt.title("Distribución del Área Total en Kilómetros Cuadrados");
```

```python
#No consigo juntar los outliers en una misma flecha

plt.scatter(df["area_water_km2"], df["population_total"], color="#20c41d", alpha=0.5)

plt.xlabel("Área de agua de la ciudad (Km$^2$)", fontdict=font)
plt.ylabel("Población", fontdict=font)
plt.title("Población por kilometro cuadrado de agua")

plt.xlim(-10, 140)
plt.xticks([0, 20, 40, 60, 80, 100, 120])
plt.ylim(-500000, 4000000)
plt.yticks([0, 500000, 1000000, 1500000, 2000000, 2500000, 3000000, 3500000, 4000000])

outliers = df[(df['population_total'] > 3500000) | (df['area_water_km2'] > 100)]

# Añadir flecha para cada outlier
for index, row in outliers.iterrows():
    plt.annotate('Outlier', xy=(row['area_water_km2'], row['population_total']),
                 xytext=(row['area_water_km2'] -50, row['population_total'] -500000),
                 arrowprops=dict(facecolor='red', arrowstyle='->'))
```

```python

altura_ciudades = df['elevation_m'].dropna()
plt.boxplot(altura_ciudades, vert = False,boxprops= {"facecolor": "red", "color": "black"}, 
            patch_artist= True, whis = 3)
plt.xlabel("Altitud (m)", fontdict=font)
plt.title("Diagrama de caja de la altitud de las ciudades de California", fontdict=font)
plt.xlim(-10, 2500)
i = 0
plt.xticks([0,500,1000,1500,2000,2500]);
```

```python

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Obtener las ciudades más pobladas de California
ciudades_mas_pobladas = df.sort_values(by='population_total', ascending=False).head(5)
colores_pastel = ['#4b54b3', '#999996', '#731e2b', '#4bb34b', '#4f3827']

#Diagrama de barras de las ciudades más pobladas
axs[0, 0].bar(ciudades_mas_pobladas['city'], ciudades_mas_pobladas['population_total'], color=colores_pastel)
axs[0, 0].set_title("Ciudades más pobladas de California")
axs[0, 0].set_ylabel("Habitantes (miles)")
axs[0, 0].set_ylim(0, 5000000)
axs[0, 0].set_yticks([1000000, 2000000, 3000000, 4000000, 5000000])
axs[0, 0].tick_params(axis='x', rotation=45)

#Histograma de la distribución del área total
axs[0, 1].hist(df['area_total_km2'], bins=40, alpha=0.7, color="steelblue")
axs[0, 1].set_title("Distribución del Área Total en Kilómetros Cuadrados")
axs[0, 1].set_xlabel("Superficie (km2)")
axs[0, 1].set_ylabel("Número de ciudades")

#Diagrama de dispersión con outliers
axs[1, 0].scatter(df["area_water_km2"], df["population_total"], color="#20c41d", alpha=0.5)
axs[1, 0].set_title("Población por kilómetro cuadrado de agua")
axs[1, 0].set_xlabel("Área de agua de la ciudad (Km$^2$)")
axs[1, 0].set_ylabel("Población")
axs[1, 0].set_xlim(-10, 140)
axs[1, 0].set_xticks([0, 20, 40, 60, 80, 100, 120])
axs[1, 0].set_ylim(-500000, 4000000)
axs[1, 0].set_yticks([0, 500000, 1000000, 1500000, 2000000, 2500000, 3000000, 3500000, 4000000])

outliers = df[(df['population_total'] > 3500000) | (df['area_water_km2'] > 100)]
# Añadir flecha para cada outlier
for index, row in outliers.iterrows():
    axs[1, 0].annotate('Outlier', xy=(row['area_water_km2'], row['population_total']),
                      xytext=(row['area_water_km2'] - 50, row['population_total'] - 500000),
                      arrowprops=dict(facecolor='red', arrowstyle='->'))

# Boxplot de la altitud
altura_ciudades = df['elevation_m'].dropna()
axs[1, 1].boxplot(altura_ciudades, vert=False, boxprops={"facecolor": "red", "color": "black"},
                  patch_artist=True, whis=3)
axs[1, 1].set_title("Diagrama de caja de la altitud de las ciudades de California")
axs[1, 1].set_xlabel("Altitud (m)")
axs[1, 1].set_xlim(-10, 2500)
axs[1, 1].set_xticks([0, 500, 1000, 1500, 2000, 2500])

# Ajustar la disposición de los subgráficos
plt.tight_layout();

plt.savefig("combined_plot.png")
```
