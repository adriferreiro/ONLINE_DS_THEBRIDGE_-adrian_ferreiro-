```python
df["ram_gb"] =df["Ram"].str.replace("GB","").astype(int) 
df["peso"] =df["Weight"].str.replace("kg","").astype(float)
mapeo = {
    'Windows 10': 1,
    'Linux': 2,
    'No OS': 3,
    'Windows 7': 4,
    'Chrome OS': 5,
    'macOS': 6,
    'Windows 10 S': 7,
    'Mac OS X': 8,
    'Android': 9
}

# Aplicar el mapeo a la columna 'OpSys'
df['OpSys_numerico'] = df['OpSys'].map(mapeo)
```

```python
screen_resolution_df = df['ScreenResolution'].copy()

# Extraer características relevantes de la resolución de la pantalla
screen_resolution_df = screen_resolution_df.str.extract(r'(IPS|Retina|Touchscreen|Full HD|Quad HD|4K Ultra HD|2560x1440|1920x1080|1366x768)', expand=False).fillna('Other')
screen_resolution_df = screen_resolution_df.str.replace(' ', '_')  # Reemplazar espacios por guiones bajos

# Asignar un valor numérico a cada característica de la pantalla
mapping = {
    'IPS': 1,
    'Retina': 2,
    'Touchscreen': 3,
    'Full_HD': 4,
    'Quad_HD': 5,
    '4K_Ultra_HD': 6,
    '2560x1440': 7,
    '1920x1080': 8,
    '1366x768': 9,
    'Other': 10
}
screen_resolution_df = screen_resolution_df.map(mapping)

# Convertir el tipo de almacenamiento a variables dummy
screen_resolution_df = pd.get_dummies(screen_resolution_df, prefix='ScreenResolution')

# Concatenar el nuevo DataFrame con el DataFrame original
df = pd.concat([df, screen_resolution_df], axis=1)

# Eliminar la columna original 'ScreenResolution'
df = df.drop(columns=['ScreenResolution'])

# Verificar el resultado
df.head()
```

```python
# Crear un nuevo DataFrame para el procesamiento
memory_df = df['Memory'].copy()

# Extraer la capacidad de almacenamiento y convertirla a GB
memory_df = memory_df.str.replace('GB', '').str.replace('TB', '000').str.extract(r'(\d+)').astype(float)
memory_df = memory_df.fillna(0)  # Rellenar valores NaN con 0
memory_df = memory_df.rename(columns={0: 'Capacity_GB'})

# Asignar un valor numérico a cada tipo de almacenamiento
memory_df['Storage_Type'] = df['Memory'].str.extract(r'(SSD|HDD|Flash Storage)').fillna('Other')

# Convertir el tipo de almacenamiento a variables dummy
memory_df = pd.get_dummies(memory_df, columns=['Storage_Type'])

# Concatenar el nuevo DataFrame con el DataFrame original
df = pd.concat([df, memory_df], axis=1)

# Eliminar la columna original 'Memory'
df = df.drop(columns=['Memory'])

# Verificar el resultado
df.head()
```
